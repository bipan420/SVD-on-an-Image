```python
from matplotlib import pyplot as plt
from matplotlib.image import imread
import numpy as np
import os
plt.rcParams['figure.figsize'] = [16,8]
```

```python
A = imread(os.path.join('dog.jpg'))
print(A.shape)
#converting the image to grayscale: ie, find mean of the image.ie all (red+green+blue)/3,
#note the axis is -1 which represents last value of shape; channel
X = np.mean(A,2)
img = plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U,S,Vt = np.linalg.svd(X,full_matrices=False)
S = np.diag(S)
j = 0
for r in (5,20,50):
    Xapprox = U[:,:r]@S[:r,:r]@Vt[:r,:]
    plt.figure(j+1)
    j += 1
    img = plt.imshow(Xapprox)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('r ='+str(r))
    plt.show()
```

```python
plt.figure(1)
#make a plot with log scalling in the y-axis
plt.semilogy(np.diag(S))
plt.title('Singular values')
plt.show()

plt.figure(2)
plt.plot(np.cumsum(np.diag(S))/np.sum(np.diag(S)))
plt.title('Singular values')
plt.show()
```
